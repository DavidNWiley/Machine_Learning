```julia
using Plots
theme(:dark)
```

```julia
function linearly_sep_data(n)
    data_1 = []
    data_2 = []
    coeff = rand(2)
    X = rand(n,2)
    
    for i = 1:n
        
        if coeff[1]*X[i,1] + coeff[2] < X[i,2]
            push!(data_1, (X[i,1], X[i,2]))
        end
        
        if coeff[1]*X[i,1] + coeff[2] > X[i,2]
            push!(data_2, (X[i,1], X[i,2]))
        end
        
    end
    
    return data_1, data_2, coeff
end
```

```julia
above, below, coeff = linearly_sep_data(250)
```

```julia
scatter([x for x in above], label = "Above")
scatter!([x for x in below], label = "Below")
plot!(x->coeff[1]*x + coeff[2], label = "f(x)")
```
